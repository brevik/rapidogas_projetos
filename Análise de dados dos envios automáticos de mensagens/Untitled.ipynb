```python
import pandas as pd
import os
```

```python
#listar nomes de arquivos de mesma extensão
def listarNomeArquivos(caminhoArquivos, extensaoArquivo):
    listaArquivos = os.listdir(caminhoArquivos) #lista todos arquivos da pasta
    nomesArquivos = []
    for l in listaArquivos:
        if extensaoArquivo in l: #lista arquivos de determinada extensão
            nomesArquivos.append(caminhoArquivos + "\\" + l)
           
    return nomesArquivos
```

```python
#função para buscar número inválido para envio e quantidade de falhas
def buscarNumeroinvalido(texto):
    nomeOcorrencia = "falha"
    qnt = 0
    listaNumeroInvalido = []
    for linha in texto:
        if nomeOcorrencia in linha:
            posicao = linha.find(" 55") #posição do caractere do início do número
            listaNumeroInvalido.append(linha[posicao+1:]) #armazenar na lista esse número
            qnt +=1
    
    return listaNumeroInvalido, qnt
```

```python
#ler arquivo
def lerArquivo(nomeArquivo):
    with open(nomeArquivo, "r", encoding="utf-8") as arq:
        texto = arq.readlines()
    
    return texto

#escrever arquivo
def escreverArquivo(nomeArquivo, modoEscrita, texto):
    with open(nomeArquivo, modoEscrita, encoding="utf-8") as arq:
        arq.write(texto)

```
